```python
import pandas as pd
import numpy as np
import os
```

```python
tune_src = r"C:\Users\labuser\ray_results\_tune_2023-06-22_12-42-13"
all_tune_files = [x for x in os.listdir(tune_src) if not x.endswith(".json")]
all_tune_files
```

```python
all_tune_files_full_src = [os.path.join(tune_src,x) for x in all_tune_files]
all_tune_files_full_src
```

```python
all_df = pd.DataFrame(columns=["name","max_map_b","max_map_b_epoch"])
for idx, src_file in enumerate(all_tune_files_full_src):
    if "progress.csv" in os.listdir(src_file):
        name = src_file
        csv_file_src = os.path.join(src_file,"progress.csv")
        progress_df = pd.read_csv(csv_file_src)
        df1 = progress_df.groupby("metrics/mAP50-95(B)").max()
        max_map_b = df1.index[-1]
        max_map_b_epoch = df1.iloc[-1].epoch
        all_df.at[idx,"name"] = name
        all_df.at[idx,"max_map_b"] = max_map_b
        all_df.at[idx, "max_map_b_epoch"]= max_map_b_epoch
    else:
        continue
```

```python
all_df['max_map_b'] = all_df['max_map_b'].apply(lambda x: int(x))
```

```python
top_five_max = all_df.nlargest(5, 'max_map_b')

# Retrieve the values from columns 'B' and 'C' for the top five rows
top_five_max_epoch= top_five_max['max_map_b_epoch']
top_five_max_name = top_five_max['name']

# Print the results
print(top_five_max_epoch)
print(top_five_max_name)

```
